```python
import pandas as pd
import csv
import plotly
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
def see_candles(tv_daily_data, title):
    #TradingView Daily Data (our ML inputs)
    #Strig you want to name the chart
    fig = go.Figure(data=go.Ohlc(x=tv_daily_data['time'],
                    open=tv_daily_data['open'],
                    high=tv_daily_data['high'],
                    low=tv_daily_data['low'],
                    close=tv_daily_data['close'])
        )
    fig.update_layout(
        xaxis=dict(fixedrange=False),  
        yaxis=dict(fixedrange=False),
        title = title

        )
    fig.show()
    return()    

```

```python
#All dataframes are on the candle distribution unless otherwise stated. 
cu_df = pd.read_csv("Relevant Data/copper.csv")
#copper
au_df = pd.read_csv("Relevant Data/gold.csv")
#gold
ag_df = pd.read_csv("Relevant Data/silver.csv")
#silver
oil_df = pd.read_csv("Relevant Data/oil.csv")
ti_ppi_df = pd.read_csv("Relevant Data/Ti_PPI.csv")
#Titanium PPI (A monthly metric)
```

## Price Charts of input variables

```python
see_candles(cu_df, "Copper Daily Price Chart")
see_candles(au_df, "Gold Daily Price Chart")
see_candles(ag_df, "Silver Daily Price Chart")
see_candles(oil_df, "Oil Daily Price Chart")
```
